```python
import matplotlib.pyplot as plt
import numpy as np

def analazyer(num, confusion_matrix):
  conf_mat = confusion_matrix

  # confusion matrix 시각화
  plt.figure()
  plt.imshow(conf_mat, interpolation='nearest', cmap=plt.cm.Blues)
  plt.title(f'Confusion matrix of model {num}')
  plt.xlabel('True label')
  plt.ylabel('Predicted label')
  plt.colorbar()

  # 축에 T와 F 표기
  tick_marks = np.arange(2)
  plt.xticks(tick_marks, ['T', 'F'])
  plt.yticks(tick_marks, ['P', 'N'])

  # 숫자 표시
  for i in range(len(conf_mat)):
      for j in range(len(conf_mat[0])):
          plt.text(j, i, conf_mat[i, j], ha='center', va='center', color='red')

  plt.show()
```

```python
conf = np.array([[1457, 126], [94, 1489]])
analazyer("Ensemble 2.3", conf)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 모델 이름
model_names = ['Model 2', 'Model Ensemble 1.2', 'Model Ensemble 2.2']

# 성능 메트릭 값들 (예시 데이터)
metrics = ['Accuracy', 'Precision', 'Recall', 'F1 Score']
values_model1 = [1,1,1,1]
values_model2 = [accuracies[7]/accuracies[1], precisions[7]/precisions[1], recalls[7]/recalls[1], f1_scores[7]/f1_scores[1]]
values_model3 = [accuracies[11]/accuracies[1], precisions[11]/precisions[1], recalls[11]/recalls[1], f1_scores[11]/f1_scores[1]]

print(values_model1)
print(values_model2)
print(values_model3)

# 각 메트릭 값들의 각도 (레이더 차트의 각도)
angles = np.linspace(0, 2*np.pi, len(metrics), endpoint=False)

# 원형 레이더 차트 생성
fig, ax = plt.subplots(figsize=(6, 6), subplot_kw={'polar': True})
ax.set_theta_offset(np.pi/2)  # 90도 회전 (12시 방향을 위로)
ax.set_theta_direction(-1)    # 시계 반대방향

# 레이더 차트의 각 메트릭 값들 표시
ax.plot(angles, values_model1, label=model_names[0])
ax.fill(angles, values_model1, alpha=0.25)
ax.plot(angles, values_model2, label=model_names[1])
ax.fill(angles, values_model2, alpha=0.25)
ax.plot(angles, values_model3, label=model_names[2])
ax.fill(angles, values_model3, alpha=0.25)

# 각도 위치에 메트릭 이름 표시
ax.set_xticks(angles)
ax.set_xticklabels(metrics)

# y축 범위 설정
ax.set_ylim(0.99, 1.01)

# 범례 추가
ax.legend()

# 그래프 제목 설정
plt.title('Model Performance Comparison')

# 그래프 표시
plt.show()
```
